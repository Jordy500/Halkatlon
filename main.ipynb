```python
# Importations ncessaires
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split, RandomizedSearchCV, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from scipy.stats import randint
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt

# Chargement des donnes
df = pd.read_csv('database.csv')

# Suppression des colonnes inutiles
df.drop(['ID', 'Z_Revenue', 'Z_CostContact', 'Complain', 'Dt_Customer'], axis=1, inplace=True)

# Suppression des lignes avec valeurs manquantes pour 'Income'
dt_clear = df.dropna(subset=['Income'])

# Dfinition des variables explicatives et de la cible
y = dt_clear['Response']
X = dt_clear.drop('Response', axis=1)
X_categorical = X.select_dtypes(exclude=[np.number])
X_encoded = pd.get_dummies(X_categorical, drop_first=True)
X_numeric = X.select_dtypes(include=[np.number])
X_final = pd.concat([X_numeric, X_encoded], axis=1)

le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Encodage des variables catgorielles
# Remplacez 'Education' et 'Marital_Status' par les noms exacts de vos variables catgorielles :


# Sparation des donnes en jeu d'entranement et jeu de test
X_train, X_test, y_train_encoded, y_test_encoded = train_test_split(X_final, y_encoded, test_size=0.2, random_state=42)


# Modle de rgression logistique (baseline)
model = LogisticRegression()
model.fit(X_train, y_train_encoded)
print("=== Rgression Logistique ===")
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
print("Prcision :", accuracy_score(y_test, y_pred))

# Dfinir la distribution des hyperparamtres pour RandomForest
param_dist = {
    'n_estimators': randint(50, 200),          # Nombre d'arbres
    'max_depth': [None, 10, 20, 30],           # Profondeur maximale
    'min_samples_split': randint(2, 10),       # Nombre min d'chantillons pour diviser un nud
    'min_samples_leaf': randint(1, 10)         # Nombre min d'chantillons dans une feuille
}

# Initialisation du modle RandomForest
rf = RandomForestClassifier(random_state=42)

# RandomizedSearchCV pour l'optimisation des hyperparamtres
random_search = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=50,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    random_state=42,
    verbose=2
)

# Entranement avec RandomizedSearchCV
random_search.fit(X_train, y_train)

# Affichage des meilleurs hyperparamtres et de la prcision
print("=== Optimisation RandomForest ===")
print("Meilleurs hyperparamtres :", random_search.best_params_)
print("Meilleure prcision :", random_search.best_score_)

# Prdictions avec le meilleur modle trouv
y_pred_rf = random_search.best_estimator_.predict(X_test)
print(classification_report(y_test, y_pred_rf))
print("Prcision RandomForest :", accuracy_score(y_test, y_pred_rf))

# Dfinir les paramtres  optimiser pour GridSearchCV
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialiser GridSearchCV
grid_search = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)

# Entraner avec GridSearchCV
grid_search.fit(X_train, y_train)

print("=== Optimisation RandomForest avec GridSearchCV ===")
print("Meilleurs hyperparamtres :", grid_search.best_params_)
print("Meilleure prcision :", grid_search.best_score_)

y_pred_rf_gs = grid_search.best_estimator_.predict(X_test)
print(classification_report(y_test, y_pred_rf_gs))
print("Prcision RandomForest avec GridSearchCV :", accuracy_score(y_test, y_pred_rf_gs))
# Exporter la base de donnes modifie vers un fichier CSV
dt_clear.to_csv('new_data.csv', index=False)
print("Base exporte avec succs sous le nom 'new_data.csv'.")





def plot_feature_importance(rf, feature_names, top_n=10):
    # Obtenir les coefficients d'importance
    importance = rf.feature_importances_
    
    # Trouver les indices des n meilleures caractristiques
    idx_top_pairs = np.argsort(importance)[::-1][:top_n]
    
    # Nommer les caractristiques
    names = [feature_names[i] for i in idx_top_pairs]

    # Crer un graphique de barres
    plt.figure(figsize=(12, 8))
    bars = plt.bar(range(top_n), importance[idx_top_pairs], color='skyblue')
        
    # Ajouter les noms des caractristiques comme tiquettes
    plt.xticks(range(top_n), names, rotation=45, ha='right')
    
    # Ajouter un titre et une lgende
    plt.title(f'Importance des caractristiques (Top {top_n})')
    plt.xlabel('Caractristiques')
    plt.ylabel('Importance')
    
    # Afficher le graphique
    plt.tight_layout()
    plt.show()

# Utilisation du code
feature_names = dt_clear.columns.tolist()

# Appliquer la fonction  votre modle RandomForestClassifier
plot_feature_importance(random_search.best_estimator_, feature_names)


# Si vous voulez utiliser toutes les caractristiques :
# Remplacez 'target_column' par le nom rel de votre colonne cible
y = dt_clear['Income']
X = dt_clear.drop(['Income'], axis=1)

# Si vous voulez utiliser uniquement les top_n caractristiques les plus importantes :
top_n_features = feature_names[:10]  # Par exemple, prendre les 10 premires caractristiques
X = dt_clear[top_n_features]
y = dt_clear['Income']

# Encoder la variable cible

# Diviser les donnes en donnes d'entranement et de test
X_train, X_test, y_train_encoded, y_test_encoded = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Initialiser le modle Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Entraner le modle
model.fit(X_train, y_train_encoded)

# Prdire sur les donnes de test
y_pred = model.predict(X_test)

# valuer le modle
accuracy = accuracy_score(y_test_encoded, y_pred)
print(f"Score de prcision : {accuracy:.3f}")
print("\nRapport de classification:")

# Afficher le rapport de classification
print("\nRapport de classification:")
print(classification_report(y_test_encoded, y_pred))

# Afficher la matrice de confusion
print("\nMatrice de confusion:")
print(confusion_matrix(y_test_encoded, y_pred))

# Affichage des informations sur le DataFrame
print(df.info())
print(df.describe())
print(df.head())
print(dt_clear.columns)
print(dt_clear.head())
print(dt_clear.info())

```
